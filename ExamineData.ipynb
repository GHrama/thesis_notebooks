```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('data_thesis.csv', skiprows=[1], encoding='utf-8-sig')
df = df.drop(df.columns[[76,75,74,73]], axis=1)
df = df.drop(df.columns[0:10], axis=1)

contexts_number_users = df.count(axis=0)[11:26]

names = ['Education','Entertainment','Finance','Gaming','Health/Fitness','Medical','Music','News','Productivity','Shopping','Social Networking','Transportation','Travel','Utility','Weather']
fig, ax = plt.subplots()
fig.canvas.set_window_title("Number of Users using the Contexts")
pos = np.arange(len(names)) + 0.5
ax.yaxis.grid(True, linestyle='--', which='major',
                   color='grey', alpha=.25)
ax.bar(pos, contexts_number_users, align='center', color='m')
ax.set_xticks(range(0,15,1))
ax.set_xticklabels(names, rotation = 60)
ax.set_ylabel('Number of Users')
plt.show()
```

```python
def basicBarPlot(title, values,label, x_labels, y_label, x_label_rotation, x_grid=False, y_grid=False, color_bar='m', xlim=None, ylim=None, file_name=None):
    if(len(values) != len(x_labels)):
        print ("Lenght of data does not match length of labels given!!")
    else:
        fig, ax = plt.subplots()
        fig.canvas.set_window_title(title)
        pos = np.arange(len(names)) + 0.5
        ax.yaxis.grid(y_grid, linestyle='--', which='major',
                           color='grey', alpha=.25)
        ax.xaxis.grid(x_grid, linestyle='--', which='major',
                           color='grey', alpha=.25)
        pos = np.arange(len(x_labels)) + 0.5
        ax.bar(pos, values, align='center', color=color_bar, label=label)
        ax.set_xticks(range(0,len(values),1))
        ax.set_xticklabels(x_labels, rotation = x_label_rotation)
        ax.set_ylabel(y_label)
        
        # xlim and ylim are lists
        if(xlim!=None):
            ax.set_xlim(xlim)
        if(ylim!=None):
            ax.set_ylim(ylim)
        
        plt.legend( loc='upper left', numpoints = 1 )
        
        # save file if name given as eps
        if(file_name != None):
            plt.savefig(file_name, format='eps', dpi=1000,bbox_inches='tight')
        
        plt.show()
        
    
```

```python
basicBarPlot("Number of Users using the Contexts", contexts_number_users,'Contexts', names, 'Number of Users', 60, y_grid=True, color_bar='g',xlim=[0,15],file_name='no_users_vs_contexts_used_static_survey.eps')
```
